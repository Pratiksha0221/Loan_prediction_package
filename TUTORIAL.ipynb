In this tutorial, we will walk through the Loan Prediction Package designed to predict whether a loan application will be approved based on various features. This package covers data loading, preprocessing, training a machine learning model, and visualizing the results.

## Step 1: Importing the Package

First, we need to import the necessary modules from the package.



```python
# Import the required modules from the package
from loan_prediction import load_data, DataProcessor, ModelTrainer, Visualizer
```

In the code above:

load_data: loads the dataset.

DataProcessor: handles the preprocessing of the dataset.

ModelTrainer: trains the machine learning model and evaluates its performance.

Visualizer: visualizes the data and model results.

## Step 2: Loading the Data

We will load the loan dataset using the load_data function.

```python
# Loading the dataset
file_path = r"C:\Users\Pratiksha Pingle\Documents\GitHub\loan_prediction\data\bankloan.csv"
loan_data = load_data(file_path)

# Data Head
loan_data.head()
```

Explanation:

The load_data function takes the path to a CSV file as an argument and returns a Pandas DataFrame containing the loan data.

In this example, loan_data.head() will display the first few rows of the loaded dataset.

## Step 3: Preprocessing the Data

The next step is preprocessing the data. We will use the DataProcessor class to clean and prepare the data for model training.

```python
# Creating an instance of the DataProcessor class
data_processor = DataProcessor(loan_data)

# Preprocessing the data
processed_data = data_processor.preprocess()
processed_data.head()
```

Explanation:

The DataProcessor class handles tasks like removing unnecessary columns and filling missing values.

The preprocess method applies these transformations to the data and returns the cleaned dataset.

## Step 4: Visualizing the Data

Before training the model, it is useful to explore the dataset visually. We can use the Visualizer class to generate various plots.

```python
# Creating an instance of the Visualizer class
visualizer = Visualizer(processed_data)

# Visualizing the data
visualizer.visualize_data()
```

Explanation:

The Visualizer class is used to generate different types of plots to explore the dataset.

The visualize_data method generates histograms for each feature in the dataset to help identify trends and distributions.

## Step 5: Training the Model

Next, we will train a Random Forest model using the ModelTrainer class.

```python
# Creating an instance of ModelTrainer
model_trainer = ModelTrainer(processed_data)

# Splitting the data into training and testing sets
model_trainer.split_data(test_size=0.25, random_state=0)

# Training the Random Forest model
model_trainer.train_model(n_estimators=100, random_state=0)
```

Explanation:

The ModelTrainer class handles training the machine learning model.

The split_data method divides the data into training and testing sets.

The train_model method trains the Random Forest model using the training set.


## Step 6: Evaluating the Model

After training the model, we can evaluate its performance by calculating various metrics like AUC and MSE.

```python
# Model Evaluation
metrics = model_trainer.evaluate_model()

# Evaluation metrics
print(metrics)
```

Explanation:

The evaluate_model method computes performance metrics like AUC (Area Under the Curve) and MSE (Mean Squared Error).

The metrics are returned as a dictionary and printed to the console.

The results indicate that:

- **AUC Score (0.9427)**: This value measures the model's ability to distinguish between the two classes (loan approval or denial). A score close to 1 indicates excellent performance, meaning the model is highly effective at distinguishing between positive and negative outcomes.
  
- **Mean Squared Error (0.0128)**: This value measures the average squared difference between the predicted and actual values. A lower MSE indicates that the model's predictions are very close to the actual values, showing good accuracy in the predictions.
